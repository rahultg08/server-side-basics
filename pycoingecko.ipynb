```python
from pycoingecko import CoinGeckoAPI
```

```python
cg = CoinGeckoAPI()
```

```python
bitcoin_data = cg.get_coin_market_chart_by_id(id = 'bitcoin', vs_currency = 'usd', days = 30)
```

```python
cg
```

```python
bitcoin_data
# gives bitcoin data in USD for past 30 days
# includes price, market cap, total volumes at the unit timestamp
```

```python
bitcoin_data_price = bitcoin_data['prices']
```

```python
import pandas as pd
```

```python
data = pd.DataFrame(bitcoin_data_price, columns = ['Timestamp', 'Price'])
```

```python
data
```

```python
data['Date'] = pd.to_datetime(data['Timestamp'], unit = 'ms')
```

```python
data
```

```python
data.Date.dt
```

```python
candlestick_data = data.groupby(data.Date.dt.date).agg({'Price': ['min', 'max', 'first', 'last']})
```

```python
candlestick_data
```

```python
import plotly.graph_objects as go
```

```python
# CandleStick Chart
fig = go.Figure(data = [go.Candlestick(x = candlestick_data.index, open = candlestick_data['Price']['first'], high = candlestick_data['Price']['max'], low = candlestick_data['Price']['min'], close = candlestick_data['Price']['last'])])
```

```python
fig.show()
```

```python
fig.update_layout(xaxis_rangeslider_visible = False, xaxis_title = 'Date', yaxis_title = 'Price (USD $)', title = 'Bitcoin Candlestick Chart over Past 30 days')
```
